# Results from transformer-based models

```python
path = "logs/LongForecasting/Transformers/"
files = os.listdir(path)

print(len(files))
data = []


for file in files:
#     print("File:",file)
    # (model)_(dataset)_(seq_len)_(pred_len)_(seed)_(seed_value)_.log
    aux = re.search("^(.*)_(.*)_(.*)_(.*)_(.*)_(.*)\..*$",file)
    model = aux.group(1)
    dataset = aux.group(2)
    seq_len = aux.group(3)
    pred_len = aux.group(4)
    seed_value = aux.group(6)
    with open(path+file, "r") as f:
        read_data = f.readlines()
        
    # (mse)_(mse_value)_(mae)_(mae_value)
    metrics = re.search("^mse:(.*), mae:(.*)\n$",read_data[-1])
    mse = metrics.group(1)
    mae = metrics.group(2)
#     print("Strings detected",model,dataset,seq_len,pred_len,seed_value,mse,mae)
    data.append([model,dataset,seq_len,pred_len,seed_value,mse,mae])

    
df_former = pd.DataFrame(data, columns=['Model', 'Dataset', 'Input_length', 'Prediction_length', 'Seed', 'MSE', 'MAE'])
df_former["Model"] = df_former["Model"].astype("string")
df_former["Dataset"] = df_former["Dataset"].astype("string")
df_former["Input_length"] = df_former["Input_length"].astype(int)
df_former["Prediction_length"] = df_former["Prediction_length"].astype(int)
df_former["Seed"] = df_former["Seed"].astype(int)
df_former["MSE"] = df_former["MSE"].astype(float)
df_former["MAE"] = df_former["MAE"].astype(float)
```

```python
# Check dataframe unique values 
for unique_v in df_former.columns[:-2]:
    print(df_former[unique_v].unique())
display(df_former[:])
```

# Stationarity

```python
# ADF: Critical value (test statistic) of the data
# p-valjue: how likely it is that your data would have occurred under the null hypothesis. Probability that null hypothesis will not be rejected
# 1%, 5%, 10%: T values

# Null hypothesis: there is a unit root, hence non Stationarity exists in the series.
# Alternative Hypothesis: Stationarity exists in the series

# If critical value (test statistic) is greater than t-values, null hypothesis cannot be rejected, indicating non-stationarity
# If p-value greater than 0.05 (95% confidence interval), null hypothesis cannot be rejected, indicanting non-stationarity

non_stat=0
list_non_stat=[]
dataset = datasets["ETTh1"].iloc[:, 1:]
for series in dataset:
    result = adfuller(datasets["ETTh1"][series])
    print("Series ", series)
    print('\tADF Statistic: %f' % result[0])
    print('\tp-value: %f' % result[1])
    print('\tCritical Values:')
    for key, value in result[4].items():
        print('\t\t%s: %.3f' % (key, value))
    if (result[0]>result[4]["5%"] and result[1]>0.05):
        print("\tNon-stationarity. Null hypothesis cannot be rejected with 95% confidence interval")
        list_non_stat.append(series)
        non_stat+=1
        
print("Number of non stationary series within {} dataset: {} out of {}".format("ETTh1",non_stat, len(dataset.columns)))

```
